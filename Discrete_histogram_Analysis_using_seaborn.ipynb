```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Use seaborn data set
tips = sns.load_dataset("tips")
print(tips.head)
```

```python
sns.histplot(data=tips, x="total_bill")
```

```python
sns.histplot(data=tips, y="total_bill")
```

```python
sns.histplot(data=tips, x="total_bill", binwidth=3)
```

```python
sns.histplot(data=tips, x="total_bill", binwidth=30)
```

```python
sns.histplot(data=tips, x="total_bill", binwidth=15)
```

Add a kernal Density, estimate to smooth the histogram, providing the complimentry infomation about the shape of distribution.  

```python
sns.histplot(data=tips, x="total_bill", kde=True)
```

```python
penguins = sns.load_dataset("penguins")
print(penguins.head)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm")
```

```python
sns.histplot(data=penguins, x="flipper_length_mm" , binwidth= 3)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm" , binwidth= 30)
#if neither x or y level is assigned the data set is treated as wideform and histogram is drawn for each numeric column.
```

```python
sns.histplot(data=penguins, x="flipper_length_mm" , kde=True)
```

We can draw multiple histogram from a long form data set with hue mapping.

```python
sns.histplot(data=penguins)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm" , hue="species")
```

The default approach to plot multiple distribution is to "layer" them but you can also "stack" them.

```python
sns.histplot(data=penguins, x="flipper_length_mm" , multiple="stack")
```

Overlapping bars can be hard to visually resolved
diffrent approach to draw a straight function.

```python
sns.histplot(data=penguins, x="flipper_length_mm" , element="step")
```
