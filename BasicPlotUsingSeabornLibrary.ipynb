```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars_data=pd.read_csv('/Toyota.csv')
```

```python
cars_data.head()
```

```python
cars_data.info()
```

```python
cars_data.dropna(axis=0)
```

```python
cars_data.dropna(axis=0,inplace= True)
```

```python
cars_data.info()
```

```python
sns.set(style="darkgrid")
sns.regplot(x=cars_data['Age'], y=cars_data['Price'])
```

```python
sns.set(style="darkgrid")
sns.regplot(x=cars_data['Age'],fit_reg=False, y=cars_data['Price'])
```

```python
sns.regplot(x=cars_data['Age'], fit_reg=False, marker="*", y=cars_data['Price'])
```

```python
sns.scatterplot(x=cars_data['Age'], y=cars_data['Price'], hue=cars_data['FuelType'], marker="*")
```

```python
sns.lmplot(x='Age', y='Price', hue='FuelType', fit_reg=False, data=cars_data, scatter_kws={"marker": "*"})
```

```python
sns.lmplot(x='Age', y='Price', hue='FuelType', fit_reg=False, legend=True, palette='Set1', data=cars_data, scatter_kws={"marker": "*"})
```

```python
sns.histplot(cars_data['Age'])
```

```python
sns.distplot(cars_data['Age'])
```

```python
sns.distplot(cars_data['Age'], kde=False)
```

```python
sns.distplot(cars_data['Age'], kde=False, bins=5)
```

```python
sns.countplot(x="FuelType",data=cars_data)
```

```python
sns.countplot(x="FuelType",data=cars_data, hue="Automatic")
```

```python
sns.boxplot(y=cars_data["Price"])
```

```python
sns.boxplot(x=cars_data["Price"])
```

```python
sns.boxplot(y=cars_data["Price"],x=cars_data["FuelType"])
```

```python
sns.boxplot(y=cars_data["Price"],x=cars_data["FuelType"],hue="Automatic", data=cars_data)
```
